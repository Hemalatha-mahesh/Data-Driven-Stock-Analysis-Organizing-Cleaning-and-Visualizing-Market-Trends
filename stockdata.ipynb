```python
#converting 2023-10 data yaml file in to csv data file
import yaml
import pandas as pd
import glob
import os

# Folder containing YAML files
folder_path = "d:/data/2023-10/"
output_file = "D:/stock/stocks_data_october_2023.csv"

all_data = []

# Loop through all YAML files in the folder
for file in glob.glob(os.path.join(folder_path, "*.yaml")):
    with open(file, "r") as f:
        data = yaml.safe_load(f)
        if isinstance(data, list):  
            all_data.extend(data)

# Convert all collected data into a single DataFrame
df = pd.DataFrame(all_data)

# Organize by Ticker 
if "Ticker" in df.columns:
    df = df.sort_values(by="Ticker")

# Save to CSV
df.to_csv(output_file, index=False)

print(f"Combined data from all YAML files saved to {output_file}")
print(df.head())
```

```python
#stocks sorted by yearly return#
import pandas as pd

# Load CSV file
file_path = "D:/stock/stock analysis/final_combined_data.csv"   
data = pd.read_csv(file_path)

# Convert 'date' column to datetime
data["date"] = pd.to_datetime(data["date"])

# Filter only 2023 and 2024
data = data[(data["date"].dt.year >= 2023) & (data["date"].dt.year <= 2024)]

# Sort by Ticker and Date
data = data.sort_values(["Ticker", "date"])

# Calculate combined return across 2023–2024
yearly_return = data.groupby("Ticker").agg(
    Open=("open", "first"),
    Close=("close", "last")
).reset_index()

# Create return column (match the name exactly)
yearly_return["YearlyReturn_2023_2024"] = ((yearly_return["Close"] - yearly_return["Open"]) / yearly_return["Open"])*100 

# Sort by return descending
sorted_stocks = yearly_return.sort_values(by="YearlyReturn_2023_2024", ascending=False)

print("Stocks sorted by YearlyReturn_2023–2024:")
print(sorted_stocks[['Ticker','YearlyReturn_2023_2024']])

# Save to CSV
#sorted_stocks.to_csv(r"D:/stock/stock analysis/Yearly_return_2023_2024.csv", index=False)
print("CSV file created successfully!")
```

```python
# 1. Volatility Analysis #

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("D:/stock/stock analysis/final_combined_data.csv")
# datetime format & sort the data 
data['date'] = pd.to_datetime(data['date'])
data = data.sort_values(by=['Ticker', 'date'])

# Calculate Daily Returns 
data['Daily_Return'] = data.groupby('Ticker')['close'].pct_change()

# Filter data from Oct 2023 to Sep 2024 
start_date = pd.Timestamp("2023-10-01")
end_date   = pd.Timestamp("2024-09-30")

data_range = data[(data['date'] >= start_date) & (data['date'] <= end_date)]

# Drop rows with null values
data_range = data_range.dropna(subset=['Daily_Return'])

# Calculate Volatility (std dev of daily returns in %) 
volatility = (
    data_range.groupby('Ticker')['Daily_Return']
    .std()
    .reset_index()
    .rename(columns={'Daily_Return': 'Volatility'})
)

# Convert to percentage
volatility['Volatility_%'] = (volatility['Volatility'] * 100).round(2)

# Find Top 10 Most Volatile Stocks 
top10_volatility = volatility.sort_values(by='Volatility', ascending=False).head(10)

# save csv file
top10_volatility.to_csv("D:/stock/stock analysis/top10_volatility.csv",index=False)
print("top10 volatility csv file saved successfully")

# Summary Table 
print(f"\n Top 10 Most Volatile Stocks ({start_date.date()} - {end_date.date()}):")
print(top10_volatility[['Ticker', 'Volatility_%']])

# Visualization
plt.figure(figsize=(12, 7))
plt.barh(top10_volatility['Ticker'], top10_volatility['Volatility_%'], color='royalblue')
plt.title(f"Top 10 Most Volatile Stocks ({start_date.date()} - {end_date.date()})", fontsize=14)
plt.xlabel("Volatility (%)")
plt.ylabel("Stock Ticker")
plt.gca().invert_yaxis()  # Show highest volatility at the top
plt.grid(axis='x', linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Cumulative Return for Top 5 Stocks

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Load Data
data = pd.read_csv("D:/stock/stock analysis/final_combined_data.csv")

# Ensure the Date column is datetime type
data['date'] = pd.to_datetime(data['date'])

# Calculate Daily Returns
# Sort by Ticker and Date
data = data.sort_values(['Ticker', 'date'])

# Daily return: percentage change
data['DailyReturn'] = data.groupby('Ticker')['close'].pct_change()

# Fill NaN values (first row for each stock)
data['DailyReturn'] = data['DailyReturn'].fillna(0)

# Calculate Cumulative Returns
# Cumulative return = product of (1 + daily return) - 1
data['CumulativeReturn'] = (1 + data['DailyReturn']).groupby(data['Ticker']).cumprod() - 1

# Identify Top 5 Performing Stocks
# Get last cumulative return for each stock
final_returns = data.groupby('Ticker')['CumulativeReturn'].last().sort_values(ascending=False)

stocks = final_returns.index.tolist()

# Filter Data for Top 5 Stocks
cumulative_return = data[data['Ticker'].isin(stocks)]
display(cumulative_return)

top5_data = final_returns.head(5).index.tolist()
top5_stocks = data[data['Ticker'].isin(top5_data)]

# save
#cumulative_return.to_csv("D:/stock/stock analysis/Cumulative_return.csv",index=False) 
print("Cumaltive return csv file saved successfully")

# Plot Cumulative Returns
plt.figure(figsize=(12,6))
sns.lineplot(data=top5_stocks, x='date', y='CumulativeReturn', hue='Ticker', marker='o')
plt.title('Cumulative Return for Top 5 Performing Stocks (Year)')
plt.xlabel('Stock over the year')
plt.ylabel('Cumulative Return')
plt.legend(title='Ticker')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 4. Stock Price Correlation #
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("D:/stock/stock analysis/final_combined_data.csv")

# Convert date to datetime
data['date'] = pd.to_datetime(data['date'])

# Sort data
data = data.sort_values(by=['Ticker', 'date'])

# Calculate daily percentage change in close price for each stock
data['DailyReturn'] = data.groupby('Ticker')['close'].pct_change() * 100

# Replace NaN (first day of each stock) with 0
data['DailyReturn'] = data['DailyReturn'].fillna(0)

# Pivot so each column is one stock, each row is a date
pivot_returns = data.pivot(index='date', columns='Ticker', values='DailyReturn')

# Compute correlation matrix
corr_matrix = pivot_returns.corr()

# Display correlation matrix
print("Correlation Matrix:")
display(corr_matrix)

# Sort and get top 10 tickers
top10 = corr_matrix.sort_values(by='Ticker', ascending=False).head(10)

# Get their names
top10_tickers = corr_matrix.index[:10] 

# Then filter original corr_matrix for both rows and columns
top10_corr_matrix = corr_matrix.loc[top10_tickers, top10_tickers]
display(top10_corr_matrix)

# save csv file
top10_corr_matrix.to_csv("D:/stock/stock analysis/top10_corr_matrix.csv")
print("top10 corr matrix saved successfully")

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Between Closing Percentage of Different Stocks")
plt.show()
```

```python
# 5.Top 5 Gainers and Losers (Month-wise) #

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("D:/stock/stock analysis/final_combined_data.csv")

# Convert 'date' to datetime
data['date'] = pd.to_datetime(data['date'])

# Sort data by Ticker and date
data = data.sort_values(by=['Ticker', 'date'])

# Extract month in YYYY-MM format
data['month'] = data['date'].dt.to_period('M').astype(str)

# Calculate monthly return
# Group by Ticker and month, take first and last closing price of the month
monthly = data.groupby(['Ticker', 'month'], as_index=False).agg(
    open_month=('close', 'first'),
    close_month=('close', 'last')
)

# Monthly return in percentage
monthly['MonthlyReturn'] = ((monthly['close_month'] - monthly['open_month']) / monthly['open_month']) * 100

#save monthly return
monthly[['Ticker','month','MonthlyReturn']].to_csv("D:/stock/stock analysis/montly_return.csv",index=False)
print("monthly return csv file saved successfully")

# Identify Top 5 Gainers & Losers for each month
# Use group_keys=False to avoid FutureWarning
top_gainers = monthly.groupby('month', group_keys=False).apply(lambda x: x.nlargest(5, 'MonthlyReturn'))
top_losers = monthly.groupby('month', group_keys=False).apply(lambda x: x.nsmallest(5, 'MonthlyReturn'))

# Optional: Check results
print("Top 5 Gainers for each month:")
print(top_gainers.head(10))
print("\nTop 5 Losers for each month:")
print(top_losers.head(10))

# Visualization
months = sorted(monthly['month'].unique())
num_months = len(months)

# Create subplots: one row per month, 2 columns (Gainers | Losers)
fig, axes = plt.subplots(num_months, 2, figsize=(14, num_months * 3))
fig.suptitle("Top 5 Gainers and Losers by Month", fontsize=16, fontweight='bold')

for i, month in enumerate(months):
    # Gainers
    g = top_gainers[top_gainers['month'] == month]
    sns.barplot(x='MonthlyReturn', y='Ticker', data=g, ax=axes[i, 0], color='green')
    axes[i, 0].set_title(f"{month} - Top 5 Gainers")
    axes[i, 0].set_xlabel("Monthly Return (%)")
    axes[i, 0].set_ylabel("")
    
    # Losers
    l = top_losers[top_losers['month'] == month]
    sns.barplot(x='MonthlyReturn', y='Ticker', data=l, ax=axes[i, 1], color='red')
    axes[i, 1].set_title(f"{month} - Top 5 Losers")
    axes[i, 1].set_xlabel("Monthly Return (%)")
    axes[i, 1].set_ylabel("")

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()

# Save results
top_gainers.to_csv("D:/stock/stock analysis/top_gainers_monthly.csv", index=False)
top_losers.to_csv("D:/stock/stock analysis/top_losers_monthly.csv", index=False)
```

```python
import pandas as pd

daily_data = pd.read_csv("D:/stock/stock analysis/Cumulative_return.csv")  
monthly_data = pd.read_csv("D:/stock/stock analysis/montly_return.csv")  
company_sector = pd.read_csv("D:/stock/stock analysis/sector_ticker_details.csv") 
yearly_data = pd.read_csv("D:/stock/stock analysis/Yearly_return_2023_2024.csv")  

# Prepare daily data
daily_data['date'] = pd.to_datetime(daily_data['date'])
daily_data['month'] = daily_data['date'].dt.to_period('M')

# Rename daily return column if needed
if 'DailyReturn' in daily_data.columns:
    daily_data.rename(columns={'DailyReturn': 'daily_return'}, inplace=True)

#  Merge monthly return
# Ensure 'month' in monthly_data is period[M]
monthly_data['month'] = pd.to_datetime(monthly_data['month']).dt.to_period('M')
daily_data = pd.merge(daily_data, monthly_data[['Ticker', 'month', 'MonthlyReturn']], 
                      on=['Ticker', 'month'], how='left')
daily_data.rename(columns={'MonthlyReturn': 'monthly_return'}, inplace=True)

# Merge company and sector info
# 
daily_data = pd.merge(daily_data, company_sector[['Ticker', 'COMPANY', 'sector']], 
                      on='Ticker', how='left')
daily_data.rename(columns={'COMPANY': 'company'}, inplace=True)

# Merge yearly return
daily_data = pd.merge(daily_data, yearly_data[['Ticker', 'YearlyReturn_2023_2024']], 
                      on='Ticker', how='left')
daily_data.rename(columns={'YearlyReturn_2023_2024': 'yearly_return'}, inplace=True)


# Reorder columns
cols = ['Ticker', 'close', 'date', 'high', 'low', 'month', 'open', 'volume', 'year',
        'daily_return', 'company', 'sector', 'yearly_return', 'monthly_return']

daily_data = daily_data[cols]

# Save final dataset
#daily_data.to_csv("D:/stock/stock analysis/final_combined_data_with_AllColumns.csv", index=False)
print("Final combined dataset saved successfully!")
display(daily_data)
```
